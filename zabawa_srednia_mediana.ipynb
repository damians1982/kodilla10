```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
lista1 = [1,2,3,4,5,6,3,3,2,2,4,5,6,2,8,9,0,40,90]
np_lista = np.array(lista1)
np_lista.shape
```

```python
plt.hist(np_lista,bins=20)
plt.show()
```

```python
mean_val = np.round(np.mean(np_lista))
mean_val
```

```python
plt.hist(np_lista, bins=20)
plt.axvline(x = mean_val, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.show()
```

```python
median_val = np.median(np_lista)
print(median_val)
```

```python
plt.hist(np_lista, bins=20)
plt.axvline(x = mean_val, color='red', label = 'Średnia')
plt.axvline(x = median_val, color='green', label = 'Mediana')
plt.legend(loc='upper right')
plt.show()
```

```python
mode_val = stats.mode(np_lista)
print(mode_val)
```

```python
plt.hist(np_lista, bins=20)
plt.axvline(x = mean_val, color='red', label = 'Średnia')
plt.axvline(x = median_val, color='green', label = 'Mediana')
plt.axvline(x = mode_val[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.show()
```
